```python
import boto3
import pandas as pd

s3 = boto3.client('s3')
```

```python

s3 = boto3.resource(
    service_name='s3',
    region_name = 'ap-south-1',
    aws_access_key_id = '',
    aws_secret_access_key = ''
)
```

```python
# Print out bucket names
for bucket in s3.buckets.all():
    print(bucket.name)
```
